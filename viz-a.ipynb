# Import packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
sns.set()
sns.get_dataset_names()
```

# Plot Plantes Dataset

```python
planets_df=sns.load_dataset('planets')
planets_df
```

```python
sns.relplot(x='year', y='distance', data=planets_df)
```

# Mass and distance are in a proportional relationship vs year

```python
sns.relplot(x="year",y="distance", hue="mass", data=planets_df)
```

```python
year_sums = planets_df.groupby("year", as_index=False).sum()
sns.relplot(x='year', y='distance', data=year_sums)
```

# Predictions data
# it appears ditance will be increased in the following years, that is correct universe is expanding faster than light

```python
sns.lmplot(x="year",y="distance",data=year_sums)
```
